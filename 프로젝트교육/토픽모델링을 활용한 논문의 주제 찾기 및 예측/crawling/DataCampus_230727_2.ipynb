```python
data = []

for i in file_list:
  if '##' in i:
    f=open("/content/drive/MyDrive/Hongik/kimmi/" +i, 'r', encoding = 'utf-8-sig')
    lines = f.readlines()
    temp = ('\n').join(lines).split('###')
    temp = [i for i in temp if len(i)>30]
    data.extend(temp)
  else:
    f=open("/content/drive/MyDrive/Hongik/kimmi/" +i, 'r', encoding = 'utf-8-sig')
    lines = f.readlines()
    temp = ('\n').join(lines)
    temp = [i for i in temp if len(i)>30]
    data.extend(temp)
```

```python
#정규표현식 활용
import re
regex = []
for i in range(len(data)):
  text = re.sub('[^0-9ㄱ-힣]', ' ', str(data[i]))
  text = re.sub(' +', ' ', text)
  regex.append(text)
```

```python
#okt 활용해서 명사추출
noun = []
for j in range(len(regex)):
  noun.append(okt.nouns(regex[j]))
  print(str(j) + "번째 수행중")
```

```python
#LDA(Latent Dirichlet Allocation) 활용 gensim모델 불러오기
import gensim

#warning 문구 삭제
import warnings
warnings.filterwarnings('ignore')
```

```python
#사전 만들기
id2word = gensim.corpora.Dictionary(noun)
```

```python
texts = noun.copy()
```

```python
corpus = [id2word.doc2bow(text) for text in texts]
```

```python
print(corpus[1:2])
```

```python
#LDA 모델 구축

#제일 먼저 topic의 개수를 몇개로 할 것인지 지정
#corpus = [id2word.doc2bow(text) for text in texts]
number = 5
```

```python
#corpus = [id2word.doc2bow(text) for text in texts]
#number = 5
model = gensim.models.ldamodel.LdaModel(
    corpus = corpus,
    id2word = id2word,
    num_topics = number,
    random_state = 1025
)
```

```python
model.show_topics(formatted = False)
```

```python
#gensim으로 부터 Coherence model 불러오기
from gensim.models.coherencemodel import CoherenceModel
coherence_model = CoherenceModel(model = model, texts = texts, dictionary = id2word, coherence = 'c_v')
```

#Coherence Model

Coherence같은 경우 주제의 일관성을 측정하는데
해당 토픽모델이 모델링이 잘 되었는지를 살펴볼 수있습니다.

토픽개수가 3개인데 0.832
토픽개수가 6개인데 0.917

*해당 토픽이 중심을 잡고 있는지
*일관성이 있는 단어들로 포함이 되어있는지

```python
#coherence_model.get_cogerence로 하나만 구하는 것도 가능, 반복적으로 확인해야 하니까 List로 정의
coherence_value = []
coherence_value.append(coherence_model.get_coherence())
```

```python
coherence_value
```

```python
start = 3
end = 10
```

```python
#모델을 여러개 만들어서 각각의 coherence score를 측정하는 형태로

for number in range(start, end):
  model = gensim.models.ldamodel.LdaModel(
          corpus = corpus,
          id2word = id2word,
          num_topics = number,
          random_state = 2500
          )

  coherence_model = CoherenceModel(model = model, texts = texts, dictionary = id2word, coherence = 'c_v')
  coherence_value.append(coherence_model.get_coherence())
```

```python
coherence_value
```

```python
#coherence value를 시각화

import matplotlib.pyplot as plt

x = range(start, end)

plt.plot(x, coherence_value)
plt.xlabel("Num Topics")
plt.ylabel("Coherence score")
plt.legend(("coherence_values"), loc='best')
plt.show()
```
